```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
class TrainingArgs:
    def __init__(self):
        pass
# args = TrainingArgs()
```

```python
ae_adam = torch.load('../checkpoints/ae-adam.pt', map_location=torch.device('cpu'))
ae_sgd = torch.load('../checkpoints/ae-sgd.pt', map_location=torch.device('cpu'))
ae_nesterov = torch.load('../checkpoints/ae-sgd-nesterov.pt', map_location=torch.device('cpu'))


adam_his = ae_adam['history']
sgd_his = ae_sgd['history']
nesterov_his = ae_nesterov['history']
```

```python
x_axis = np.arange(len(adam_his['train']))

fig, axs = plt.subplots(1,2,figsize=(18,6))
axs[0].plot(x_axis, adam_his['train'], label='Adam')
axs[0].plot(x_axis, sgd_his['train'], label='Sgd')
axs[0].plot(x_axis, nesterov_his['train'], label='Sgd-Nesterov')
axs[0].legend(['Adam', 'SGD', 'SGD-Nesterov'])
axs[0].set_title("Training reconstruction loss (learning rate = 1e-2)")
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')

axs[1].plot(x_axis, adam_his['val'], label='Adam')
axs[1].plot(x_axis, sgd_his['val'], label='Sgd')
axs[1].plot(x_axis, nesterov_his['val'], label='Sgd-Nesterov')
axs[1].legend(['Adam', 'SGD', 'SGD-Nesterov'])
axs[1].set_title("Validation reconstruction loss (learning rate = 1e-2)")
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Loss')

plt.tight_layout()
plt.show()
```
